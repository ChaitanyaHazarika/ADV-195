# Task 1 - Find the diabetic and hight blood pressure patients across all age groups, and conclude the risk heart stroke is higher in which age group

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt 

#read the csv
df = pd.read_csv('heart_disease.csv')
df
```

```python
#Filter by diabetes(condition will be who has diabetes) and create new dataframe
diabetes= df.loc[df['diabetes'] == 1] 
diabetes
```

```python
#On this new data frame perform group operation as per age and create new dataframe 
grpbyagediabetes= diabetes.groupby('age')['diabetes'].count().reset_index()
grpbyagediabetes
```

```python
#Filter by high_blood_pressure(condition will be who has high_blood_pressure) and create new dataframe
patienthighbp= df.loc[df['high_blood_pressure'] == 1]
patienthighbp
```

```python
#On this new data frame perform group operation as per age and create new dataframe 
grpbyagebp= patienthighbp.groupby('age')['high_blood_pressure'].count().reset_index()
grpbyagebp
```

```python
#plot the scatter graph to show which age group is more prone to diabetes
plt.figure(figsize=(12,8))
diabetes1=grpbyagediabetes['diabetes']

age= grpbyagediabetes['age']
plt.scatter(age, diabetes1, label= "Diabitic Patient")

age2= grpbyagebp['age']
bp= grpbyagebp['high_blood_pressure']
plt.scatter(age2,bp, label="Patient With Hight BP")
plt.xlabel("Age Group")
plt.ylabel("Diabitic And high BP")
plt.legend()
```

# Task 2 - Find as per gender who has not normal platelets level in blood

```python
#Filter by platelets(condition lesser then 150000 OR greater then 450000) and create new dataframe
platelets= df.loc[(df['platelets'] <= 150000.00)| (df['platelets'] >= 450000.00)]
platelets
```

```python
#On this new dataframe perform group operation as per gender and create new dataframe 
grpbygender= platelets.groupby('gender')['platelets'].count().reset_index()
grpbygender


```

```python
#Plot a pie chart as per the gender to show the percentage of male and female who has not normal platelets
value= grpbygender['platelets']
label= grpbygender['gender']

plt.pie(value, labels= label, autopct= "%0.1F%%", radius= 2)
plt.show()
```

Conclusion - The percentage of males having not normal platelets in blood are higher than the femalessssss
